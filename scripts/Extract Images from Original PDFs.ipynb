```python
import os
import subprocess
```

```python
basepath = "/data1/dyang/Projects/Evan_CNNRetrieval/score-retrieval/data/pdfs"
outdir = "/data1/dyang/Projects/Evan_CNNRetrieval/score-retrieval/data/ref_jpgs"
if not os.path.exists(outdir):
    os.mkdir(outdir)
for i in os.listdir(basepath):
    print(i)
    fname = os.path.join(basepath,i)
    pieceName = os.path.splitext(i)[0]
    outpath = outdir+"/"+pieceName+"/"+pieceName+"/"+"page.jpg"
    if not os.path.exists(os.path.dirname(outpath)):
        os.makedirs(os.path.dirname(outpath))
    subprocess.call(['convert','-limit','memory','32GiB','-limit','map','8GiB','-limit','disk','16GiB','-density', '300', '-alpha', 'remove', '-resize', '2550', fname, outpath])
```
